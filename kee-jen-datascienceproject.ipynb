```python
import pandas as pd

df = pd.read_csv("/kaggle/input/data-science-job-posting-on-glassdoor/Uncleaned_DS_jobs.csv")
df.head()
```

```python
df.shape       # rows and columns
df.info()      # column details
df.head(10)    # first 10 rows
```

```python
df['Salary Estimate'].eq("-1").any()
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape 
```

```python
salary = df['Salary Estimate'].apply(lambda x:x.split('(')[0])
```

```python
minus_kd = salary.apply(lambda x: x.replace('K', '').replace('$', ''))
```

```python
print( minus_kd)
```

```python
def get_min(x):
    parts = x.split('-')
    return int(parts[0])  # works for both 1 or 2 parts

def get_max(x):
    parts = x.split('-')
    if len(parts) == 2:   # if range exists
        return int(parts[1])
    else:                 # single value case
        return int(parts[0])

df['min_salary'] = minus_kd.apply(get_min)
df['max_salary'] = minus_kd.apply(get_max)
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
```

```python
df.head(10)
```

```python
df['company_name'] = df['Company Name'].apply(lambda x: x.split('\\')[0])
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1].lstrip() if ',' in x else None)
```

```python
df.head(10)
```

```python
df.job_state.value_counts()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis =1 )
```

```python
df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2025-x)
```

```python
df.head(10)
```

```python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.head(10)
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head(10)
```

```python
df.to_csv('Cleaned_Data', index= False)
```
